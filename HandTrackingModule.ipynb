```python
import cv2
import mediapipe as mp
import time
import math
import numpy as np
from pynput.keyboard import Controller, Key
```

```python
draw_line = False
start_point = None
end_point = None

last_length = None
last_angle = None

minHand = 50
maxHand = 300

angle = 0
angleBar = 400 
angleDeg = 0
minAngle = -65.25
maxAngle = 0.0

keybord = Controller()

mphands = mp.solutions.hands
hands = mphands.Hands(max_num_hands = 1, min_detection_confidence = 0.5)
mpdraw = mp.solutions.drawing_utils

cap = cv2.VideoCapture(0)


while True:
    ret, img = cap.read()

    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results = hands.process(imgRGB)

    if results.multi_hand_landmarks:
        for handLMS in results.multi_hand_landmarks:
            for id, lm in enumerate(handLMS.landmark):
                h, w, c = img.shape
                cx, cy = int(lm.x * w), int(lm.y * h)
                if id == 4 or id == 8:
                    if id == 4:
                        start_point = (cx, cy)
                    elif id == 8:
                        end_point = (cx, cy)
                        draw_line = True

                    cv2.circle(img, (cx, cy), 15, (255, 0, 255), cv2.FILLED)

            mpdraw.draw_landmarks(img, handLMS, mphands.HAND_CONNECTIONS)

    if draw_line and start_point and end_point:
        cv2.line(img, start_point, end_point, (0, 0, 255), 3)

        center_point = ((start_point[0] + end_point[0]) // 2, (start_point[1] + end_point[1]) // 2)
        cv2.circle(img, center_point, 15, (255, 0, 255), cv2.FILLED)

        length = math.hypot(end_point[0] - start_point[0], end_point[1] - start_point[1])
        #print(length)
        angle = np.interp(length, [minHand, maxHand],[minAngle, maxAngle])
        angleBar = np.interp(length, [minHand, maxHand],[400, 150])
        angleDeg = np.interp(length, [minHand, maxHand],[0, 100])
        #print(int(length), angle)

        if last_length:
            if length > last_length:
                keybord.press(Key.media_volume_up)
                keybord.release(Key.media_volume_up)
                print('VOL UP')
            elif length < last_length:
                keybord.press(Key.media_volume_down)
                keybord.release(Key.media_volume_down)
                print('VOL DOWN')  
                
        last_angle = angle
        last_length = length

        if length < 50 :
            cv2.circle(img, center_point, 15, (0,255,0), cv2.FILLED)

    cv2.rectangle(img, (50,150), (85,400),(255,0,0), 3)
    cv2.rectangle(img, (50, int(angleBar) ),(85, 400), (255,0,0), cv2.FILLED)
    cv2.putText(img, f'{int(angleDeg)}%', (40, 120), cv2.FONT_HERSHEY_COMPLEX, 2, (0,9,255), 3)
    cv2.imshow('Frame', img)
    
    key = cv2.waitKey(1)
    if key == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
